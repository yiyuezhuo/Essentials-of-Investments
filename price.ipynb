# 投资学精要笔记2

## 单因素模型与beta值

为了运用前面指出的求出最优风险投资组合的方法。我们必须知道市面上所有资产的收益的期望向量和协方差矩阵，如果说前者是一个常用的指标的话。如果有10000个资产，则协方差矩阵需要1000000000个数来表示相关信息。这是在计算和数据获取上都是不能被接受的。因此，我们建立更简单的模型来表示风险。

一个显然的想法是，既然所有组合比起来过于复杂，我们可以让它们与特定的资产组合进行比较，然后在需要协方差矩阵的时候(不是与特定资产比较的时候)利用这个关系反推其值。这个方法显然忽略了一些信息，就像矩阵的矩阵分解方法一样，我们可以拿掉一些维来让信息丢失并不是那么严重。

单因素模型将证券的超额收益分解为预期超额收益率，宏观因素决定的超额收益率与随机因素决定的超额收益率。

$$
R = a + \beta R_M + e \\
$$
其中$R$为超额收益率，$R_M$为市场的超额收益率，e为公司内在的超额收益率或随机误差。

我们可以通过考察两个单位历时比较长的时间收益的时间序列来直接回归出$a$与$\beta$值。我们将$\beta$看成是总体市场对收益的影响力，换言之，风险（可变收益）的来历之一。$e$则是另一个，回归产生的另一个有益的统计量是$e$方差的估计值。

一般来说，使用指数（如标普500）当$R_M$，这些值的确一定意义上反应总体市场，也易于获得。

## 非系统风险的消除

超额收益中来自市场的风险叫做系统风险，$e$对应的是非系统风险。可以想象如果我们不断地分散投资组合到一定程度上独立的各个投资组合上，我们就可以将非系统风险不断降低乃至忽略不计。于是在单元素模型中我们只需要考虑单元素表示的系统风险。

注意这一点暗示了各个具体的证券之间的关系被看成被系统因素完全决定了，这显然不是真的，不过也许我们能够容忍这种偏离。

## 资本资产定价模型

资本资产定价模型通过假设所有投资人都是一样的（除了允许风险偏好不一样外），于是他们选择的都是一样的有效边界上的最优风险投资组合，至多只是在与无风险资产的分配上不同。但无论如何，他们都持有相同的夏普比率和（有风险资产内）相同分配比例的投资组合。

如果从中移出博弈的成分，那么似乎的确应该有某种最优策略，于是每个人都选择被称为市场组合的策略，十分稳定。然而现实中，由于不断出现的错误，使得在博弈上利用这些错误的收益比稳态下才有最高收益的市场组合更高。于是市场组合，如果真有这么个东西的话，从来就没有成为常态。常态是无数同时发生的愚蠢与套利的反复循环而已。

无论如何，现在我们相信现实的证券在套入这个模型后产生的指标，如“（模型告诉你的）预期回报率”似乎有一种魔力，会使现实的确向这个预测值移动。于是我们就可以利用这一点去识别那些转瞬即逝的微小偏差进行投机套利，同时这类行为本身也为回归提供了动力（难道不是一种自我实现吗，因为一帮人根据这个模型判断低估而买入，而的确发生了上升来让低估的预测实现，这和庄家有什么区别？当然某种意义上，资本资产定价模型就是假设这种行为能光速精准完成来建立理论的。所以它的流行使它的假设也的确越来越靠得住脚了）

### 没有截距与beta值

资本资产定价模型最重要的结论是，个别资产的风险溢价（本身收益/利率减无风险收益/利率）与市场组合溢价成正比。而且当市场组合溢价为0时个别资产溢价也为0.即

$$
E(r_d) - r_f = \beta_0 (E(r_M)-r_f) \\
$$

即没有截距项。而$\beta_0$则反映了“成正比”中的比例系数。即市场风险的大小。

没有截距是因为一旦突然有了更高的收益，表现在此模型上一个额外的正截距$a$来解释左边超额收益的突然增长。市场组合显然将会变化来以更高的比例持有超额收益的证券，于是$E(r_M) - r_f$超额收益本身这一项上升，而由于比例升高，$\beta_0$也在上升，于是项$\beta_0 (E(r_M)-r_f)$上升，在超额收益不变情况下，$a$当然就会下降，直到降到$0$为止。

模型中证券的非系统风险不在考虑中，因为市场中的投资者都被假设通过充分分散而将非系统风险降到了忽略不计的程度。所以他们只在系统风险上进行风险-收益权衡。这就形成了这个正比例关系。

## 证券市场线与alpha值

把上面方程的左边从超额收益变成收益后，

$$
E(r_d) = r_f + \beta_0 (E(r_M)-r_f) \\
$$

所对应的直线就是*证券市场线*，为模型根据秘制来源的$\beta$与秘制来源的市场组合的超额收益率为单个资产预测的“公平”的收益率。如果我们有一个这样的“公平收益率”，却发现实际收益率或“实际”预测收益率与其不相同，则后者减前者就构成了alpha值。即观测水平与“公平”水平的偏离现象。大于0说明存在正截距式超额收益率，小于0说明存在负截距式超额收益率。按照模型的思路，对于前者将引发市场组合向该资产偏移，从而使该资产价格上升，收益率本身下降，另一方面市场组合超额收益率上升，而$\beta$上升，是截距系数$\alpha$归于0.如果果真如此，那么的确就存在套利机会了。

## 单指数模型

将市场组合以某个指数（如标普500）替换，就得到了指数模型。指数模型可以通过回归直接估计出$\beta$值，在无约束模型里还可以估计$\alpha$值，再与有约束模型比较看$\alpha$值是否显著，还是说回归出这个值完全可以看成是实际上没有但因为随机因素导致的。

然而虽然指数与“市场组合”有某种含糊的共性，但是这一步跨出去的的确有些令人生疑。似乎只是形式上与资本资产定价理论类似，但实质只是个贫乏的回归模型而已。

## 多指数模型

单因素模型可以自然扩展到多因素模型，如在Fama and French(FF)三因素模型中，影响超额收益率的变量除了市场组合超额收益率以外，还有公司规模，账面价值-市场价值比率两个因素。它们无非只是扩展了线性方程的形式而已。引入一些$\beta_1$,$\beta_2$之类的。

它的指数化妥协版本用指数去表示这些因素，如公司规模可以使用小公司类的平均收益率去减大公司类的平均收益率（SMB）这个指数去表示。账面价值-市场价值也用类似的方法将证券分成两部分计算差值处理（HML）。这种方法引入了显而易见的专断，如到底如何划分这两个类。不过倒的确是个通用型很好的方法。

在其他多因素/指数模型的扩展中，我们甚至引入宏观经济变量还进行解释，总之越来越像胡乱的回归了。只不过这里很有趣的赋予截距一个特殊的地位，一般回归却只考虑偏斜率的值是否显著。

## 套利定价理论

套利定价理论指出对于允许alpha!=0的无约束模型来说，我们可以通过线性组合一些这样的资产来产生无系统风险投资组合，而又有正的期望收益率，从而使夏普比率变得无限大。这是不可能的，将立刻引发套利（其论证倒和上面的类似），而使得alpha归于0.
